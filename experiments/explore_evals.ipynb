```python
import pandas as pd
import numpy as np

csv_path = "/net/vast-storage/scratch/vast/kellislab/artliang/magneton/magneton/tests/esmc/evaluation_results.csv"
df = pd.read_csv(csv_path)
df
```

```python
# Compute per-class accuracy
def compute_per_class_accuracy(df):
    class_accuracies = {}
    for label in df['true_mappings'].unique():
        # Filter rows for the current class
        class_df = df[df['true_mappings'] == label]
        # Calculate accuracy
        correct_predictions = class_df[class_df['true_mappings'] == class_df['pred_mappings']]
        accuracy = len(correct_predictions) / len(class_df)
        class_accuracies[label] = accuracy
    return class_accuracies

class_accuracies = compute_per_class_accuracy(df)
class_accuracies
```

```python
sorted_accuracies = sorted(class_accuracies.items(), key=lambda item: item[1])
sorted_accuracies
```

```python
# TODO Fix evals so that top_k_probs is a list not a string
import ast
df['top_k_probs'] = df['top_k_probs'].apply(ast.literal_eval)

def analyze_confidence(df):
    confidences = []
    for probs in df['top_k_probs']:
        # Assuming top_k_probs is a list of lists, take the highest probability
        max_confidence = max(probs)
        confidences.append(float(max_confidence))
    return np.mean(confidences), np.std(confidences)

mean, std = analyze_confidence(df)
mean, std
```

All of the above is an example of how we can extract more metrics from the actual prediction distributions of an evaluation run for a specific model and interpro type
