```python
# Merge all EWC CSVs on 'DMS ID' from ground_truth_csv and 'assay' from each EWC CSV
result = ground_truth_csv[['DMS ID', 'ESM-C (300M)']].copy()

# List of (csv, suffix) pairs
ewc_csvs = [
    (active_ewc_csv, 'active_ewc'),
    (binding_ewc_csv, 'binding_ewc'),
    (conserved_ewc_csv, 'conserved_ewc'),
    (domain_ewc_csv, 'domain_ewc'),
    (homo_ewc_csv, 'homo_ewc'),
]

for csv, suffix in ewc_csvs:
    result = result.merge(
        csv[['assay', 'correlation']].rename(columns={'correlation': f'correlation_{suffix}', 'assay': 'DMS ID'}),
        on='DMS ID',
        how='left'
    )

# Optionally, you can save the result
# result.to_csv('correlations.csv', index=False)
```

```python
# Calculate pairwise correlations between all methods (excluding the identifier column)
correlation_matrix = result.drop(columns=['DMS ID']).corr()
print(correlation_matrix)

# Visualize the correlation matrix
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix Between Different Methods')
plt.show()

# Compare each method to ESM-C (300M) baseline
for col in result.columns[1:]:  # Skip ESM-C (300M) column
    corr = result['ESM-C (300M)'].corr(result[col])
    print(f"Correlation with ESM-C (300M): {col} = {corr:.3f}")
```
